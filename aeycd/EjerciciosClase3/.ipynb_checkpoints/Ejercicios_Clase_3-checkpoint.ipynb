````python
# %matplotlib notebook
import matplotlib.cm
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data1 = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/wtageinf.csv', encoding='utf8')
data2 = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/lenageinf.csv', encoding='utf8')
````

````python
data1.head()
````

````python
data2.head()
````

 ### División por sexo

- Niños

````python
data1_M = data1[data1.Sex == 1]
data1_M = data1_M.groupby('Agemos', as_index = True).sum()
data1_M.plot(y=['P3','P5','P10','P25', 'P50', 'P75', 'P90', 'P95','P97'],
        kind='line', 
        title=u'Percentiles de Niños Peso por Edad', 
        figsize=(14,10),
        colormap=matplotlib.cm.get_cmap('viridis'))
plt.ylabel('Peso')
plt.show()
````

- Niñas

````python
data1_F = data1[data1.Sex == 2]
data1_F = data1_F.groupby('Agemos', as_index = True).sum()
data1_F.plot(y=['P3','P5','P10','P25', 'P50', 'P75', 'P90', 'P95','P97'],
        kind='line', 
        title=u'Percentiles de Niñas Peso por Edad', 
        figsize=(14,10),
        colormap=matplotlib.cm.get_cmap('viridis'))
plt.ylabel('Peso')
plt.show()
````

**Historia de la célebre Reina de España Doña Juana, llamada vulgarmente, La Loca**

**Contar palabras y ordenar por frecuencia**
- Limpiar preludio y licencia de Project Gutenberg

````python
f = open("textoplano.txt","r") #http://www.gutenberg.org/cache/epub/20099/pg20099.txt

lineas = f.readlines()  
texto = lineas[34:937] 
texto = ''.join(texto)  
````

- Omitir “palabras vacías” (stop words) y símbolos

````python
!pip install spacy
!pip install spacy download es
!pip install textacy
import textacy 
import spacy



lenguaje = textacy.text_utils.detect_language(texto) # Detectando el idioma con taxtacy
nlp = spacy.load('es')                               # defino el tipo de modelo
texto = textacy.preprocess.remove_punct(texto, marks='\n')

documento = textacy.doc.Doc(texto, metadata=None, lang= lenguaje)#acá lo defino como un texto
frecuencia = documento.to_bag_of_terms(ngrams= 1 , named_entities=True, normalize='lemma',
                weighting='count', as_strings=True, filter_stops= True, filter_punct = True)

frecuencia = pd.DataFrame.from_dict(frecuencia, orient='index')
frecuencia = frecuencia.sort_values(by=[0], ascending=False)
frecuencia[:100]
````

**Personajes Principales**

````python
personajes = textacy.extract.named_entities(documento, include_types=None, exclude_types=None, 
                                                  drop_determiners=True, min_freq=10)

personajes = pd.Series(list(personajes)).astype(str).str.replace(","," ").unique()
personajes = pd.DataFrame(personajes,columns=['personajes'])
display(personajes)
````


- Extraemos conjuntos sujeto, verbo y objeto (**SVO**)


````python
documento_g = textacy.extract.subject_verb_object_triples(documento)
documento_g = pd.DataFrame(list(documento_g), columns=['sujeto','verbo','objeto']).astype(str).replace(["(",")"],["",""])
documento_g
````
